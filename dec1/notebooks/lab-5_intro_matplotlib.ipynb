## Curves

```python
# create a numpy array of 100 equally spaced numbers between 1 and 10
x = np.linspace(0, 10, 100)
```

```python
# plot lines and curves with plt.plot
fig = plt.figure()

plt.plot(x, np.sin(x))  # sine curve
plt.show()
```

```python
# add customization via arguments
fig = plt.figure()

plt.plot(x, np.sin(x), '+', color='blue')
plt.show()
```

```python
# Two curves in the same plot
fig = plt.figure()

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

### Labeling plots

```python
# Add chart labels
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.title("Sine vs Cosine Curve")
plt.xlabel("x"); plt.ylabel("f(x)");
```

### Legends

```python
# Adding legends
fig = plt.figure()
plt.plot(x, np.sin(x), '-', label='Sine curve')
plt.plot(x, np.cos(x), '--', label='Cosine curve')

plt.title("Sine vs Cosine Curve"); plt.xlabel("x"); plt.ylabel("f(x)")
plt.legend();
```

## Displaying two figures together

```python
plt.figure()  # create a plot figure
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# Or displaying plots side-by-side
plt.figure(figsize=(18, 6))  

plt.subplot(1, 2, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x));
```

## ax Interface

For all Matplotlib plots, we start by creating a figure and an axes. In their simplest form, a figure and axes can be created as follows:

```python
# fig = plt.figure()
# ax = plt.axes()
```

In Matplotlib, the *figure* is the container that contains all the objects representing axes, graphics, text, and labels. 

The *axes* is what we see above: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization. 

Its common use the variable name `fig` to refer to a figure instance, and `ax` to refer to an axes instance or group of axes instances.

```python
# Also called the object-oriented interface
fig, ax = plt.subplots()
# fig = plt.plots()
# ax = plt.axes()

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x));
```

We can forego using ax and directly plot the graph as well using just the `plt` interface. Using `ax` will give you a object-oriented interface and can give more fine-grained control over the plots.

```python
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

# Some useful Plot types

## Boxplots / Box and Whiskers Plot

![](images/boxplot.png)

All single variable statistics in one plot

```python
np.random.seed(100)   # this makes sure we all get the same random values

# we create normalized dataset with mean=100 & sd=10
x = np.random.normal(100, 10, 200)
y = np.random.normal(50, 30, 200)

fig = plt.figure()
plt.boxplot([x, y]);
```

## Histogram

```python
fig = plt.figure()

plt.hist(x, alpha=0.5)
plt.hist(y, alpha=0.5);
```

## Bar plots (stacked)

```python
# Group two variable by Year 
year = [2014, 2015, 2016, 2017, 2018, 2019]  
val1 = [39, 117, 98, 54, 28, 15]  
val2 = [ 0,   10, 13, 56, 39, 14]  
```

```python
plt.bar(year, val1, label='val1')
plt.bar(year, val2, bottom=val1, label='val2')

plt.xlabel('Year')  
plt.ylabel('Values')
plt.legend();
```
